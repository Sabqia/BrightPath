## import libraries

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

## Loading the Dataset

```python
data = pd.read_csv("student_lifestyle_dataset.csv")
data
```

# Splitting the Stress_Level Column into Separate Columns

```python
data

# Create new columns for each stress level
data['Low_Stress_Level'] = data['Stress_Level'].apply(lambda x: 1 if x == 'Low' else 0)
data['Moderate_Stress_Level'] = data['Stress_Level'].apply(lambda x: 1 if x == 'Moderate' else 0)
data['High_Stress_Level'] = data['Stress_Level'].apply(lambda x: 1 if x == 'High' else 0)

# View the updated DataFrame
print(data.head())
```

# Deleting the Stress_Level Column

```python
# Drop the Stress_Level column
data = data.drop(columns=['Stress_Level'])

# View the updated DataFrame
print(data.head())
```

## What is the correlation between the frequency of social interactions, sleep hours per day, physical activity hours per day, and the high stress levels of students?

### Extracts the two columns, High_Stress_Level and Social_Hours_Per_Day.

```python
#Extract the 'High_Stress_Level' and 'Social_Hours_Per_Day','Sleep_Hours_Per_Day', 'Physical_Activity_Hours_Per_Day' columns
selected_columns = data[['High_Stress_Level', 'Social_Hours_Per_Day', 'Sleep_Hours_Per_Day', 'Physical_Activity_Hours_Per_Day']]
print(selected_columns)
```

## Visualizing Correlation Between Social Hours and High Stress Levels

```python
# Sample data (assuming your `data` DataFrame already exists)
selected_columns = data[['High_Stress_Level', 'Social_Hours_Per_Day', 'Sleep_Hours_Per_Day', 'Physical_Activity_Hours_Per_Day']]

# Compute correlation matrix
correlation_matrix = selected_columns.corr()

# Visualize the correlation matrix as a heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", cbar=True)
plt.title('Correlation Heatmap')
plt.show()
```

## Correlation Value:

### This heatmap reveals that High_Stress_Level has a negative correlation with both Sleep_Hours_Per_Day and Physical_Activity_Hours_Per_Day, indicating that students who sleep more or engage in more physical activity tend to experience lower stress levels. In contrast, the correlation between High_Stress_Level and Social_Hours_Per_Day is weak or negligible, suggesting that the frequency of social interactions does not significantly influence stress levels. Additionally, the relationships among the independent variables Social_Hours_Per_Day, Sleep_Hours_Per_Day, and Physical_Activity_Hours_Per_Day may show weak to moderate correlations, highlighting interdependencies between lifestyle factors, though their direct impact on stress appears limited. This suggests that managing sleep and physical activity is more critical for reducing stress than focusing on social interaction frequency.

# Question 2

# Does consistently getting less than 6 hours of sleep increase stress levels in students?

## Extract the columns of high stress level and sleep hours per day

```python
# Extract the relevant columns
df = data[['High_Stress_Level', 'Sleep_Hours_Per_Day']]
df
```

## Bar Chart Visualization and Analysis

```python
# Filter the data based on sleep hours
sleep_less_than_6 = df[df['Sleep_Hours_Per_Day'] < 6]
sleep_6_or_more = df[df['Sleep_Hours_Per_Day'] >= 6]

# Calculate the average stress level for each group
avg_stress_less_than_6 = sleep_less_than_6['High_Stress_Level'].mean()
avg_stress_6_or_more = sleep_6_or_more['High_Stress_Level'].mean()

# Prepare data for plotting
sleep_categories = ['Less than 6 hours', '6 or more hours']
avg_stress_levels = [avg_stress_less_than_6, avg_stress_6_or_more]

# Create the bar graph
plt.bar(sleep_categories, avg_stress_levels, color=['red', 'green'])

# Adding titles and labels
plt.title('Average Stress Level vs Sleep Hours')
plt.xlabel('Sleep Hours Per Day')
plt.ylabel('Average Stress Level')
plt.ylim(0, 1)  # stress level is between 0 and 1

# Show the plot
plt.show()

# Step 3: Display the results
print("Average Stress Level (less than 6 hours of sleep):", avg_stress_less_than_6)
print("Average Stress Level (6 hours or more of sleep):", avg_stress_6_or_more)

# Compare the results
if avg_stress_less_than_6 > avg_stress_6_or_more:
    print("Consistently getting less than 6 hours of sleep seems to increase stress levels.")
else:
    print("There is no clear indication that less than 6 hours of sleep increases stress levels.")

```
